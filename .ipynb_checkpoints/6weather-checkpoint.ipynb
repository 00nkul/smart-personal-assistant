```python
import requests, json 
from tkinter import *  
import tkinter as tk  
from functools import partial 
from PIL import Image, ImageTk

top = Tk()  

cityname = tk.StringVar()

top.geometry("1500x900+100+100")
top.title("Weather Room")
top.configure(background='#809fff')

# Enter your API key here 
api_key = "7df8d9ee8f32046a7c512a5fec692e5a"

# base_url variable to store url 
base_url = "http://api.openweathermap.org/data/2.5/weather?"


lbl_city= tk.Label(top,text='Jalandhar',bg='#809fff')
lbl_city.config(font=("Courier", 20))
lbl_city.place(x=550,y=200)

lbl_room1= tk.Label(top,text='Weather Room',bg='#809fff',fg="white")
lbl_room1.config(font=("Courier", 40))
lbl_room1.place(x=500,y=100)


lbl_temp1= tk.Label(top,text='Temperature',bg='#809fff')
lbl_temp1.config(font=("Courier", 20))
lbl_temp1.place(x=200,y=400)

lbl_temp2= tk.Label(top,text='Temperature',bg='#809fff',fg="white")
lbl_temp2.config(font=("Courier", 25))
lbl_temp2.place(x=200,y=450)


lbl_pressure1= tk.Label(top,text='Pressure',bg='#809fff')
lbl_pressure1.config(font=("Courier", 20))
lbl_pressure1.place(x=500,y=400)

lbl_pressure2= tk.Label(top,text='Pressure',bg='#809fff',fg="white")
lbl_pressure2.config(font=("Courier", 25))
lbl_pressure2.place(x=500,y=450)

lbl_humadity1= tk.Label(top,text='Humadity',bg='#809fff')
lbl_humadity1.config(font=("Courier", 20))
lbl_humadity1.place(x=800,y=400)

lbl_humadity2= tk.Label(top,text='Humadity',bg='#809fff',fg="white")
lbl_humadity2.config(font=("Courier", 25))
lbl_humadity2.place(x=800,y=450)

lbl_cloud1= tk.Label(top,text='Short Description',bg='#809fff')
lbl_cloud1.config(font=("Courier", 20))
lbl_cloud1.place(x=1100,y=400)

lbl_cloud2= tk.Label(top,text='Short Description',bg='#809fff',fg="white")
lbl_cloud2.config(font=("Courier", 25))
lbl_cloud2.place(x=1100,y=450)

txt_city = tk.Entry(top,textvariable=cityname)
txt_city.config(font=("Courier", 20))
txt_city.place(x=600,y=600)

image13 = Image.open("temperature.png")
photo13 = ImageTk.PhotoImage(image13)
label13 = tk.Label(width=100,height=100,image=photo13,bg="#809fff")
label13.image = photo13
label13.place(x=250,y=300)

image14 = Image.open("pressure.png")
photo14 = ImageTk.PhotoImage(image14)
label14 = tk.Label(width=100,height=100,image=photo14,bg="#809fff")
label14.image = photo14
label14.place(x=550,y=300)


image15 = Image.open("humidity.png")
photo15 = ImageTk.PhotoImage(image15)
label15 = tk.Label(width=100,height=100,image=photo15,bg="#809fff")
label15.image = photo15
label15.place(x=830,y=300)

image16 = Image.open("description.png")
photo16 = ImageTk.PhotoImage(image16)
label16 = tk.Label(width=100,height=100,image=photo16,bg="#809fff")
label16.image = photo16
label16.place(x=1200,y=300)

imagew = Image.open("rainbow.png")
photow= ImageTk.PhotoImage(imagew)
labelw = tk.Label(width=200,height=200,image=photow,bg="#809fff")
labelw.image = photow
labelw.place(x=1100,y=0)

imager = Image.open("rainbow.png")
photor = ImageTk.PhotoImage(imager)
labelr = tk.Label(width=200,height=200,image=photor,bg="#809fff")
labelr.image = photor
labelr.place(x=100,y=0)

def specific(city):
    city_name =city.get()
    lbl_city.config(text=city_name)
    complete_url = base_url + "appid=" + api_key + "&q=" + city_name 
    response = requests.get(complete_url) 
    x = response.json() 
    if x["cod"] != "404": 
        y = x["main"] 
        current_temperature = y["temp"] 
        current_pressure = y["pressure"] 
        current_humidiy = y["humidity"] 
        z = x["weather"] 
        weather_description = z[0]["description"] 

        # print following values 
        #print(" Temperature (in kelvin unit) = " +str(current_temperature) +"\n atmospheric pressure (in hPa unit) = " +str(current_pressure) +"\n humidity (in percentage) = " +str(current_humidiy) +"\n description = " +str(weather_description)) 
    
        val1=str(round(current_temperature - 273.15))+str(chr(176))+"C"
        lbl_temp2.config(text=val1)
    
        val2=str(current_pressure)+" hPa"
        lbl_pressure2.config(text=val2)
    
        val3 =str(current_humidiy)+" %"
        lbl_humadity2.config(text=val3)
    
        val4 = str(weather_description)
        lbl_cloud2.config(text=val4)    
    else: 
        print(" City Not Found ") 
        lbl_temp2.config(text="City Not Found")
        lbl_pressure2.config(text="City Not Found")
        lbl_humadity2.config(text="City Not Found")
        lbl_cloud2.config(text="City Not Found")
    

# Give city name 
city_name = "Jalandhar"
complete_url = base_url + "appid=" + api_key + "&q=" + city_name 
response = requests.get(complete_url) 
x = response.json() 
if x["cod"] != "404": 
    y = x["main"] 
    current_temperature = y["temp"] 
    current_pressure = y["pressure"] 
    current_humidiy = y["humidity"] 
    z = x["weather"] 
    weather_description = z[0]["description"] 

    # print following values 
    #print(" Temperature (in kelvin unit) = " +str(current_temperature) +"\n atmospheric pressure (in hPa unit) = " +str(current_pressure) +"\n humidity (in percentage) = " +str(current_humidiy) +"\n description = " +str(weather_description)) 
    
    val1=str(round(current_temperature - 273.15))+str(chr(176))+"C"
    lbl_temp2.config(text=val1)
    
    val2=str(current_pressure)+" hPa"
    lbl_pressure2.config(text=val2)
    
    val3 =str(current_humidiy)+" %"
    lbl_humadity2.config(text=val3)
    
    val4 = str(weather_description)
    lbl_cloud2.config(text=val4)    


specific=partial(specific,cityname)  
btncheck= tk.Button(top,text='Check',command=specific)
btncheck.config(font=("Courier", 20))
btncheck.place(x=700,y=700)


    
top.mainloop()
```

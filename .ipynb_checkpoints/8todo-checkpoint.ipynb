```python
import tkinter as tk  
from tkinter import messagebox  
from functools import partial  
from PIL import Image, ImageTk
import sqlite3
from functools import partial  

root = tk.Tk()  
root.geometry('1500x900+100+100')    
root.title('Signup PersonalChatboat')  
root.configure(background='#809fff')

lbl_heading= tk.Label(root,text='To Do List',bg='#809fff',fg='white')
lbl_heading.config(font=("Courier", 40))
lbl_heading.place(x=550,y=100)

conn = sqlite3.connect('chatboat.db')

cursor = conn.execute("SELECT * FROM TODO")

def call_delete(n1):
    val1 = (n1.get())      
    conn.execute("DELETE FROM  TODO WHERE ID='"+val1+"'")
    conn.commit()
    messagebox.showinfo("Message","Deleted!")  
     
    
def call_insert(n1,n2):
    val1 = (n1.get())  
    val2 = (n2.get())
    
    conn.execute("INSERT INTO TODO (Title,Description) VALUES ('"+val1+"', '"+val2+"')");
    conn.commit()
    messagebox.showinfo("Message","List Added !")  

y_p=200
count=0
for row in cursor:
    list_no=str(row[0])+") "+str(row[1])
    t1 =tk.StringVar(root, value=list_no)
    txt1=tk.Label(root,textvariable=t1)
    txt1.config(font=("Courier", 20),bg="#809fff")
    txt1.place(x=100,y=y_p)
     
    
    t2 = tk.StringVar(root, value=row[2])
    txt2=tk.Entry(root,textvariable=t2)
    txt2.config(font=("Courier", 20))
    txt2.place(x=100,y=50+y_p,width=800)
    
    y_p+=120
    count+=1
    
t3 =tk.StringVar(root,"Add To Do ")
txt3=tk.Label(root,textvariable=t3)
txt3.config(font=("Courier", 20),bg="#809fff")
txt3.place(x=1100,y=200)

t4 = tk.StringVar(root)
txt4=tk.Entry(root,textvariable=t4)
txt4.config(font=("Courier", 20))
txt4.place(x=1000,y=250,width=400)

t5 = tk.StringVar(root)
txt5=tk.Entry(root,textvariable=t5)
txt5.config(font=("Courier", 20))
txt5.place(x=1000,y=300,width=400)

call_insert = partial(call_insert,t4,t5)  

btncheck1 = tk.Button(root,text='Insert',command=call_insert)
btncheck1.config(font=("Courier", 20))
btncheck1.place(x=1150,y=350)

t3 =tk.StringVar(root,"Delete To Do")
txt3=tk.Label(root,textvariable=t3)
txt3.config(font=("Courier", 20),bg="#809fff")
txt3.place(x=1100,y=700)

t6 = tk.StringVar(root)
txt6=tk.Entry(root,textvariable=t6)
txt6.config(font=("Courier", 20))
txt6.place(x=1000,y=750,width=400)


call_delete = partial(call_delete,t6)  
btncheck2 = tk.Button(root,text='Delete',command=call_delete)
btncheck2.config(font=("Courier", 20))
btncheck2.place(x=1150,y=800)  

image13 = Image.open("copybook.png")
photo13 = ImageTk.PhotoImage(image13)
label13 = tk.Label(width=256,height=256,image=photo13,bg="#809fff")
label13.image = photo13
label13.place(x=1050,y=440)
 
 
    

root.mainloop()  
```

```python
import tkinter as tk
from tkinter import *
from tkinter import messagebox  
from functools import partial 
import requests
import random 
import sys
from bs4 import BeautifulSoup
import pyttsx3 
from PIL import Image, ImageTk


top = Tk()  
top.geometry("1500x900+100+100")
top.title("Chat Room")
top.configure(background='#809fff')

lbl_city= tk.Label(top,text='Send Message',bg='#809fff',fg='white')
lbl_city.config(font=("Courier", 40))
lbl_city.place(x=550,y=100)

code = tk.StringVar()
title = tk.StringVar()
description = tk.StringVar()

def insert(c1,t1,d1):
    code=str(c1.get())
    title=str(t1.get())
    description=str(d1.get())
    r = requests.get('https://jnv2jee.com/attendance/python_project/insert.php?code='+code+'&title='+title+'&description='+description)
    content = r.content.decode(encoding='UTF-8')
    soup = BeautifulSoup(r.content.decode(encoding='UTF-8'), "lxml")
    reviews = soup.find_all()
    for i in reviews:
        statusb=str(i)
    if statusb=="<p>success</p>":
        messagebox.showinfo("Success","Message Added Successfully")      
    else:    
        messagebox.showerror("Error","Please Check Your Internet Connection")  
    
code = tk.Entry(top,textvariable=code)
code.config(font=("Courier", 20))
code.place(x=600,y=400)

title = tk.Entry(top,textvariable=title)
title.config(font=("Courier", 20))
title.place(x=600,y=500)

description = tk.Entry(top,textvariable=description)
description.config(font=("Courier", 20))
description.place(x=600,y=600)

insert=partial(insert,code,title,description)  

btncheck= tk.Button(top,text='Send',command=insert)
btncheck.config(font=("Courier", 20))
btncheck.place(x=700,y=700)

image16 = Image.open("message.png")
photo16 = ImageTk.PhotoImage(image16)
label16 = tk.Label(width=200,height=200,image=photo16,bg="#809fff")
label16.image = photo16
label16.place(x=650,y=150)

lbl_city= tk.Label(top,text='Code',bg='#809fff',fg='white')
lbl_city.config(font=("Courier", 40))
lbl_city.place(x=400,y=380)

lbl_city= tk.Label(top,text='Title',bg='#809fff',fg='white')
lbl_city.config(font=("Courier", 40))
lbl_city.place(x=400,y=470)

lbl_city= tk.Label(top,text='Description',bg='#809fff',fg='white')
lbl_city.config(font=("Courier", 40))
lbl_city.place(x=200,y=580)


image17 = Image.open("girl1.png")
photo17 = ImageTk.PhotoImage(image17)
label17 = tk.Label(width=200,height=200,image=photo17,bg="#809fff")
label17.image = photo17
label17.place(x=100,y=0)

image17 = Image.open("boy1.png")
photo17 = ImageTk.PhotoImage(image17)
label17 = tk.Label(width=200,height=200,image=photo17,bg="#809fff")
label17.image = photo17
label17.place(x=1100,y=0)



top.mainloop()
```

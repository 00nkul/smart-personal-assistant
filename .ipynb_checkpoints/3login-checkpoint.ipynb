```python
import tkinter as tk  
from tkinter import messagebox  
from functools import partial  
from PIL import Image, ImageTk
import sqlite3


root = tk.Tk()  
root.geometry('900x900+700+100')    
root.title('Signup PersonalChatboat')  


conn = sqlite3.connect('chatboat.db')
 
 
   
def call_result(n1, n2):  
    val1 = (n1.get())  
    val2 = (n2.get()) 
     
    cursor = conn.execute("SELECT * from USER WHERE EMAIL='"+val1+"' AND PASSWORD='"+val2+"'")
    count_users=0
    for row in cursor:
        count_users+=1
    
    if count_users==0:
        messagebox.askretrycancel("Something went wrong","Please Try Again?")
    else:
        messagebox.showinfo("Information","Login Successfull")  
    
          
   
 
    return  
   



value1 = tk.StringVar()
value2 =tk.StringVar()

 
image3 = Image.open("ico3.png")
photo3 = ImageTk.PhotoImage(image3)
label3 = tk.Label(width=48,height=48,image=photo3)
label3.image = photo3
label3.place(x=100,y=300)

image3 = Image.open("ico4.png")
photo3 = ImageTk.PhotoImage(image3)
label3 = tk.Label(width=48,height=48,image=photo3)
label3.image = photo3
label3.place(x=100,y=400)


lbl_email = tk.Label(root,text='Email')
lbl_email.config(font=("Courier", 20))
lbl_email.place(x=180,y=300)

lbl_password = tk.Label(root,text='Password')
lbl_password.config(font=("Courier", 20))
lbl_password.place(x=180,y=400)
 
txt_email = tk.Entry(root,textvariable=value1)
txt_email.config(font=("Courier", 20))
txt_email.place(x=400,y=300)

txt_password = tk.Entry(root,textvariable=value2,show="*")
txt_password.config(font=("Courier", 20))
txt_password.place(x=400,y=400)



call_result = partial(call_result,value1,value2)  
  
buttonCal = tk.Button(root, text="Sign Up", command=call_result) 
buttonCal.config(font=("Courier", 20))
buttonCal.place(x=380,y=500)
  
root.mainloop()  
```
